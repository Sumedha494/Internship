```python
import pandas as pd

df = pd.read_excel("customers_cleaned.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.columns.tolist()
```

```python
df['first_name'] = df['first_name'].str.strip().str.lower()
df['last_name'] = df['last_name'].str.strip().str.lower()
df['company'] = df['company'].str.strip().str.lower()
df['city'] = df['city'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.lower()
```

```python
df.isnull().sum()
```

```python
df['subscription_date'] = pd.to_datetime(df['subscription_date'], errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
df.to_excel("customer_final.xlsx", index=False)
```
